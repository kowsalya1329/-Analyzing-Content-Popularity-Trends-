```python
merged = pd.merge(Reactions,Content_result,on='Content ID')
merged
```

```python
final = pd.merge(merged,ReactionTypes,on='Reaction Type')
final.head(5)
```

```python
category_scores = final.groupby('Category')['Score'].sum().reset_index()
category_scores  # Calculate total scores for each category
```

```python
top_5_categories = category_scores.sort_values(by='Score', ascending=False).head(5)
print(top_5_categories) # Sort categories by total score in descending order and get top 5
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_5_categories['Category'], top_5_categories['Score'])
plt.xlabel('Category')
plt.ylabel('Aggregate Popularity Score')
plt.title('Top 5 Categories by Aggregate Popularity Score')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Provided data
top_5_categories = {
    'Category': ['animals', 'healthy eating', 'technology', 'science', 'culture'],
    'Score': [71242, 69067, 68521, 66549, 64952]
}

# Convert data to DataFrame
import pandas as pd
df = pd.DataFrame(top_5_categories)

# Calculate percentages
df['Percentage'] = (df['Score'] / df['Score'].sum()) * 100

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Percentage'], labels=df['Category'], autopct='%1.1f%%')
plt.title('Popularity Percentage Share from Top 5 Categories')
plt.show()
```
